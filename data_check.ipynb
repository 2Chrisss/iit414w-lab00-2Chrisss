## Jolpica API Query

```python
url="https://api.jolpi.ca/ergast/f1/2024/driverstandings/"
response = requests.get(url, timeout=30)
payload = response.json()
print(f"HTTP Status Code: {response.status_code}")

data = payload['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
print(f"Number of records: {len(data)}")

print("\nFirst 3 driver entries:")
for i in range(3):
    driver_info = data[i]['Driver']
    points = data[i]['points']
    full_name = f"{driver_info['givenName']} {driver_info['familyName']}"
    nationality = driver_info['nationality']
    
    print(f" - {full_name}, {nationality}, {points} points")
```
